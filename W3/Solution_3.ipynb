## Task 1. Write code to compute the following six centrality metrics for a graph using Networkx:

Dataset: /data/karate_edgelist.txt.

Each row denotes the edge between two nodes.

Caculate centrality meassures:

1. Degree centrality
2. Betweenness centrality
3. Closeness centrality 
4. Eigenvector centrality
5. Katz centrality
6. PageRank




```python
import networkx as nx
import networkx
import matplotlib.pyplot as plt

G = nx.read_edgelist('./data/karate_edgelist.txt')
nx.draw(G, with_labels = True)
#nx.draw(G, pos=nx.spiral_layout(G))
#nx.draw(G, pos=nx.shell_layout(G),with_labels = True)
#nx.draw(G, pos=nx.spiral_layout(G),with_labels = True)
plt.show()
```

```python
#Degree centrality
nx.degree_centrality(G)
```

```python
#Betweenness centrality
nx.betweenness_centrality(G)
```

```python
#Closeness centrality
nx.closeness_centrality(G)
```

```python
#Eigenvector centrality
nx.eigenvector_centrality(G, weight='weight',max_iter = 600)
```

```python
# Katz centrality
katz_centrality = nx.katz_centrality_numpy(G, alpha=0.1, beta=1.0, weight='weight')
katz_centrality
```

```python
# PageRank Centrality
nx.pagerank(G, weight='weight')
```

### Addtional questions

1. Which node has the highest Degree centrality?
2. Which node has the highest Betweenness centrality?
3. Which node has the highest Closeness centrality?
4. Which node has the highest Eigenvector centrality?

```python
# 1. Which node has the highest Degree centrality?
degree_cent = nx.degree_centrality(G)
max_degree_node = max(degree_cent, key=degree_cent.get)
print(f"Node with highest Degree centrality: {max_degree_node} (value: {degree_cent[max_degree_node]:.4f})")

# 2. Which node has the highest Betweenness centrality?
betweenness_cent = nx.betweenness_centrality(G)
max_betweenness_node = max(betweenness_cent, key=betweenness_cent.get)
print(f"Node with highest Betweenness centrality: {max_betweenness_node} (value: {betweenness_cent[max_betweenness_node]:.4f})")

# 3. Which node has the highest Closeness centrality?
closeness_cent = nx.closeness_centrality(G)
max_closeness_node = max(closeness_cent, key=closeness_cent.get)
print(f"Node with highest Closeness centrality: {max_closeness_node} (value: {closeness_cent[max_closeness_node]:.4f})")

# 4. Which node has the highest Eigenvector centrality?
eigenvector_cent = nx.eigenvector_centrality(G, weight='weight', max_iter=600)
max_eigenvector_node = max(eigenvector_cent, key=eigenvector_cent.get)
print(f"Node with highest Eigenvector centrality: {max_eigenvector_node} (value: {eigenvector_cent[max_eigenvector_node]:.4f})")
```

## Task 2.  Construct a graph from adjacency matrix in the adj_matrix.csv

Dataset: /data/adj_matrix.csv.

The file denotes a adjacency matrix of a graph.

Caculate centrality meassures:

1. Degree centrality
2. Betweenness centrality
3. Closeness centrality 
4. Eigenvector centrality
5. Katz centrality
6. PageRank

```python
import pandas as pd
import networkx as nx
input_data = pd.read_csv('./data/adj_matrix.csv', index_col=0)
G = nx.Graph(input_data.values)
nx.draw(G)
```

```python
# Degree centrality
degree_cent = nx.degree_centrality(G)
print("Degree Centrality:", degree_cent)
```

```python
# Betweenness centrality
betweenness_cent = nx.betweenness_centrality(G)
print("Betweenness Centrality:", betweenness_cent)
```

```python
# Closeness centrality
closeness_cent = nx.closeness_centrality(G)
print("Closeness Centrality:", closeness_cent)
```

```python
# Eigenvector centrality
eigenvector_cent = nx.eigenvector_centrality(G, max_iter=600)
print("Eigenvector Centrality:", eigenvector_cent)
```

```python
# Eigenvector centrality
eigenvector_cent = nx.eigenvector_centrality(G, max_iter=600)
print("Eigenvector Centrality:", eigenvector_cent)
```

```python
# Katz centrality
katz_centrality = nx.katz_centrality_numpy(G, alpha=0.1, beta=1.0)
print("Katz Centrality:", katz_centrality)
```

```python
# PageRank Centrality
pagerank_scores = nx.pagerank(G)
print("PageRank Scores:", pagerank_scores)
```

### Addtional questions

1. Which node has the highest Degree centrality?
2. Which node has the highest Betweenness centrality?
3. Which node has the highest Closeness centrality?
4. Which node has the highest PageRank score?

```python
def top_nodes(scores):
    max_value = max(scores.values())
    return [node for node, value in scores.items() if value == max_value], max_value

deg_nodes, _ = top_nodes(degree_cent)
bet_nodes, _ = top_nodes(betweenness_cent)
clo_nodes, _ = top_nodes(closeness_cent)
pr_nodes, _ = top_nodes(pagerank_scores)

print(f"Highest Degree centrality: {', '.join(map(str, deg_nodes))}")
print(f"Highest Betweenness centrality: {', '.join(map(str, bet_nodes))}")
print(f"Highest Closeness centrality: {', '.join(map(str, clo_nodes))}")
print(f"Highest PageRank: {', '.join(map(str, pr_nodes))}")
```
